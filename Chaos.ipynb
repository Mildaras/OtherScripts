```python
import matplotlib.pyplot as plt
import numpy as np
import random

def random_point_in_triangle(v1, v2, v3):
    """
    Generate a random point inside the triangle formed by v1, v2, v3.
    """
    u = random.random()
    v = random.random()
    if u + v > 1:
        u, v = 1 - u, 1 - v
    x = v1[0] + u * (v2[0] - v1[0]) + v * (v3[0] - v1[0])
    y = v1[1] + u * (v2[1] - v1[1]) + v * (v3[1] - v1[1])
    return (x, y)

def midpoint(p1, p2):
    """
    Calculate the midpoint between two points p1 and p2.
    """
    return ((p1[0] + p2[0]) / 2, (p1[1] + p2[1]) / 2)

# Vertices of the triangle
vertex1 = (0, 0)
vertex2 = (1, 0)
vertex3 = (0.5, np.sqrt(3) / 2)
vertices = [vertex1, vertex2, vertex3]

# Initial random point inside the triangle
current_point = random_point_in_triangle(vertex1, vertex2, vertex3)

# Plotting setup
plt.figure(figsize=(6, 6))
triangle = plt.Polygon([vertex1, vertex2, vertex3], closed=True, fill=None, edgecolor='blue')
plt.gca().add_patch(triangle)

# Number of iterations to calculate new points
iterations = 500000
points_x = [current_point[0]]
points_y = [current_point[1]]

for _ in range(iterations):
    # Pick a random vertex
    random_vertex = random.choice(vertices)
    # Find the midpoint between the current point and the chosen vertex
    current_point = midpoint(current_point, random_vertex)
    # Save the new point
    points_x.append(current_point[0])
    points_y.append(current_point[1])

# Plot all points
plt.scatter(points_x, points_y, s=0.00001, color='red')

# Adjust plot limits and appearance
plt.xlim(-0.1, 1.1)
plt.ylim(-0.1, 1.1)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Chaos Game: Points in Triangle")
plt.show()
```
